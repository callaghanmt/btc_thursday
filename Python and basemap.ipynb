# Using Basemap

```python
from mpl_toolkits.basemap import Basemap
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
map = Basemap (projection='merc', lat_0=39, lon_0=29, resolution='l', area_thresh=500.0, 
               llcrnrlon= 20,
               llcrnrlat= 59,
               urcrnrlon= 32,
               urcrnrlat= 72,
               )
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color='coral')
map.drawmapboundary()

map.drawmeridians(np.arange(0,360,30))
map.drawparallels(np.arange(-90,90,30))

lat=63
lon=27
x,y = map(lon, lat)
map.plot (x,y, 'bo', markersize=12)
plt.text (x+0.25, y+0.25, 'town1')

lat=69
lon=21
x,y = map(lon, lat)
map.plot (x,y, 'rd', markersize=12)

plt.show()
```
